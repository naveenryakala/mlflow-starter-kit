```python
# ============================================================================
# STEP 1: Import All Required Libraries
# ============================================================================
# Let's import everything we need to build, train, and track our ML model

import pandas as pd                              # Data manipulation library
from sklearn import datasets                    # For loading built-in datasets
from sklearn.linear_model import LogisticRegression  # A classification algorithm
from sklearn.metrics import accuracy_score      # To measure model performance
from sklearn.model_selection import train_test_split  # To split data for training/testing
import mlflow                                  # MLflow for experiment tracking
from mlflow.models import infer_signature      # To create model signatures

print("✅ All libraries imported successfully!")
```

```python
# ============================================================================
# STEP 2: Connect to MLflow Tracking Server
# ============================================================================
# Tell MLflow where to store our experiment data
# Make sure the server is running: mlflow server --host 127.0.0.1 --port 5000

mlflow.set_tracking_uri("http://127.0.0.1:5000")

print("✅ Connected to MLflow server at http://127.0.0.1:5000")
```

```python
# ============================================================================
# STEP 3: Load Dataset and Train First Model
# ============================================================================
# The Iris dataset contains measurements of iris flowers in 4 dimensions:
# - Sepal Length, Sepal Width, Petal Length, Petal Width
# The target is to classify them into 3 species: Setosa, Versicolor, Virginica

# Load the Iris dataset
# x = features (the flower measurements)
# y = labels (which species each flower is)
x, y = datasets.load_iris(return_X_y=True)
print(f"✅ Loaded Iris dataset: {x.shape[0]} samples, {x.shape[1]} features")

# Split data into training (80%) and testing (20%) sets
# Training data: used to teach the model
# Testing data: used to evaluate how well the model learned
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
print(f"📊 Training samples: {len(x_train)}, Testing samples: {len(x_test)}")

# ============================================================================
# Define model hyperparameters (settings that control how the model learns)
# ============================================================================
# - penalty: "l2" adds regularization to prevent overfitting
# - solver: "lbfgs" is an optimization algorithm (method to learn)
# - max_iter: how many times to repeat the learning process (1000 = thorough)
# - multi_class: "auto" means let scikit-learn decide the best approach
# - random_state: 8888 = seed for reproducibility (same random numbers each time!)

params = {
    "penalty": "l2",           # Type of regularization
    "solver": "lbfgs",         # Optimization method
    "max_iter": 1000,          # Maximum iterations
    "multi_class": "auto",     # Automatic multi-class strategy
    "random_state": 8888       # For reproducibility
}

# Create and train the Logistic Regression model
lr = LogisticRegression(**params)  # Create model with our parameters
lr.fit(x_train, y_train)            # Train on the training data
print(f"✅ Model trained successfully with parameters: {params}")
```

```python
# Display the test data to see what the model will make predictions on
x_test
```

```python
# ============================================================================
# Make Predictions on Test Data
# ============================================================================
# Now use the trained model to predict what species each test flower is
# Remember: the model has NEVER seen these test flowers before (good test!)

y_pred = lr.predict(x_test)  # Predict the classes (0=Setosa, 1=Versicolor, 2=Virginica)
print("🔮 Predictions on test data:")
print(y_pred)  # Show the predictions
```

```python
# ============================================================================
# Evaluate Model Performance
# ============================================================================
# accuracy_score compares our predictions with the actual values
# It tells us: what percentage of predictions were correct?

accuracy = accuracy_score(y_test, y_pred)  # Compare predictions to actual values
print(f"Accuracy Score: {accuracy:.4f}")   # Show accuracy (0.0 = terrible, 1.0 = perfect!)
print(f"Accuracy Percentage: {accuracy*100:.2f}%")
```

## 📊 MLflow Tracking - Log Results to Dashboard

Now that we've trained and evaluated the model, let's save all that information to MLflow so we can:
- Compare this run with other runs later
- See the history of all our experiments
- Register the best model for deployment

We'll track:
1. **Parameters**: The settings we used (penalty, solver, etc.)
2. **Metrics**: How well it performed (accuracy score)  
3. **Model**: The actual trained model file (can load later!)
4. **Tags**: Notes about what this run was for


```python
# ============================================================================
# STEP 4: Log First Model Run to MLflow
# ============================================================================
# MLflow tracking lets us record every detail about this training run

# Connect to MLflow server
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")

# Create a new experiment called "MLFLOW Quickstart"
# All runs will be grouped under this experiment
mlflow.set_experiment("MLFLOW Quickstart")

# Start a new MLflow run (with statement auto-closes it)
print("📝 Starting MLflow run 1...")
with mlflow.start_run():
    # Log the hyperparameters we used during training
    # This lets us remember what settings produced what results
    mlflow.log_params(params)
    print("   ✅ Parameters logged")

    # Log the accuracy metric (our primary performance metric)
    mlflow.log_metric("accuracy", accuracy)
    print(f"   ✅ Accuracy metric logged: {accuracy:.4f}")

    # Add a human-readable note (tag) about what this run was
    # Helpful for remembering the purpose later
    mlflow.set_tag("Training", "Logistic Regression with l2 penalty on Iris dataset")
    print("   ✅ Tags added")

    # ====================================================================
    # Model Signature: Tells MLflow what data format the model expects
    # This is important for deploying the model to APIs later
    # ====================================================================
    signature = infer_signature(x_train, lr.predict(x_train))
    print("   ✅ Model signature inferred")

    # Log the trained model to MLflow
    # This saves the model so we can load and use it later without retraining!
    model_info = mlflow.sklearn.log_model(
        sk_model=lr,                              # The trained model
        artifact_path="iris_model",               # Folder name in MLflow
        signature=signature,                      # Input/output format
        input_example=x_train,                    # Example data format
        registered_model_name="tracking-quickstart",  # Name for the model registry
    )
    print("   ✅ Model logged")

print("✅ Run 1 completed and logged to MLflow!")
```

## 🔄 Train Second Model with Different Hyperparameters

Now let's train a second model with DIFFERENT settings to see if we can get better results.  
This demonstrates the key benefit of MLflow: **comparing multiple experiments easily!**

Key differences:
- **Solver**: Changed from "lbfgs" to "newton-cg" (different optimization method)
- **Random state**: Changed from 8888 to 1000 (different random seed)
- Everything else the same

This is how we find the BEST hyperparameters!

```python
# ============================================================================
# STEP 5: Train Second Model with Different Hyperparameters
# ============================================================================
# Let's try a DIFFERENT set of parameters to see if we can improve accuracy

# Define NEW model hyperparameters (changed solver and random_state)
# This is hyperparameter tuning: testing different settings to find the best ones
params2 = {
    "solver": "newton-cg",         # Different solver method (more iterations but may be more accurate)
    "max_iter": 1000,              # Same as before
    "multi_class": "auto",         # Same as before  
    "random_state": 1000           # Different random seed
}

# Create and train the second Logistic Regression model
lr2 = LogisticRegression(**params2)
lr2.fit(x_train, y_train)  # Train on same data as first model for fair comparison
print(f"✅ Model 2 trained with parameters: {params2}")
```

```python
# # ============================================================================
# Predictions from Second Model
# ============================================================================

y_pred2 = lr2.predict(x_test)  # Predictions from model 2
print("🔮 Predictions from Model 2:")
print(y_pred2)
```

```python
# ============================================================================
# Evaluate Model 2
# ============================================================================

accuracy2 = accuracy_score(y_test, y_pred2)  # Calculate accuracy for model 2
print(f"Model 2 Accuracy Score: {accuracy2:.4f}")
print(f"Model 2 Accuracy Percentage: {accuracy2*100:.2f}%")

# Compare the two models
print(f"\n📊 Model Comparison:")
print(f"   Model 1 (lbfgs): {accuracy:.4f}")
print(f"   Model 2 (newton-cg): {accuracy2:.4f}")
if accuracy2 > accuracy:
    print(f"   ✅ Model 2 is BETTER by {(accuracy2-accuracy)*100:.2f}%")
else:
    print(f"   Model 1 is better by {(accuracy-accuracy2)*100:.2f}%")
```

```python
# ============================================================================
# STEP 6: Log Second Model Run to MLflow
# ============================================================================
# Now log our second model with its different parameters
# This creates a new "run" in the same experiment for easy comparison

print("📝 Starting MLflow run 2...")
with mlflow.start_run():
    # Log parameters for model 2
    mlflow.log_params(params2)
    print("   ✅ Parameters logged")

    # Log accuracy for model 2
    mlflow.log_metric("accuracy", accuracy2)
    print(f"   ✅ Accuracy metric logged: {accuracy2:.4f}")

    # Add a tag explaining this run
    mlflow.set_tag("Training", "Logistic Regression with newton-cg solver on Iris dataset")
    print("   ✅ Tags added")

    # Create signature for model 2
    signature2 = infer_signature(x_train, lr2.predict(x_train))
    print("   ✅ Model signature inferred")

    # Log the second model
    model_info2 = mlflow.sklearn.log_model(
        sk_model=lr2,
        artifact_path="iris_model",
        signature=signature2,
        input_example=x_train,
        registered_model_name="tracking-quickstart",  # Same registry as model 1
    )
    print("   ✅ Model logged")

print("✅ Run 2 completed and logged to MLflow!")
print("\n🎉 SUCCESS! Compare both runs at: http://127.0.0.1:5000")
```
