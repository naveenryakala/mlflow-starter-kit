## MLFlow Tracking Server

```python
# Import the MLflow library to enable experiment tracking
import mlflow
```

```python
# Connect to the local MLflow tracking server running on port 5000
mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
# Create or set the active experiment where our runs will be logged
mlflow.set_experiment("check localhost connection")

# Start a new MLflow run to log metrics
with mlflow.start_run():
    # Log some dummy metrics for testing out the connection
    mlflow.log_metric("test",1)
    mlflow.log_metric("naveen",2)
```

```python
# Start another run to see how multiple runs are recorded under the same experiment
with mlflow.start_run():
    mlflow.log_metric("test-1",1)
    mlflow.log_metric("naveen-1",2)
```

```python
# Start a third run logging slightly different metrics
with mlflow.start_run():
    mlflow.log_metric("test-2",1)
    mlflow.log_metric("naveen-2",2)
```
