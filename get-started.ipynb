# 🚀 MLflow Tracking Server - Getting Started

## Welcome to MLflow!

This notebook demonstrates the **basics of MLflow**:
1. Connect to the MLflow Tracking Server
2. Create an experiment (like a project container)
3. Start runs (individual training attempts)
4. Log metrics (like accuracy, loss, etc.)

**Key Concepts:**
- **Experiment**: A container for related runs (like a project)
- **Run**: A single training attempt with its own parameters and metrics
- **Metrics**: Numbers we want to track (accuracy, loss, F1-score, etc.)
- **Parameters**: Settings we use to train the model (learning rate, epochs, etc.)

**What this notebook teaches you:**
- How to connect to a local MLflow server 
- How to organize experiments
- How to log multiple runs
- How to see results in the MLflow UI

Don't worry if you don't fully understand yet - we'll go step by step!


```python
# ============================================================================
# STEP 1: Import MLflow
# ============================================================================
# MLflow is the library that lets us track ML experiments and manage models
# We import it so we can use its functions in this notebook
import mlflow

print("✅ MLflow imported successfully!")
print(f"   MLflow version: {mlflow.__version__}")
```

```python
# ============================================================================
# STEP 2: Connect to MLflow Tracking Server
# ============================================================================
# The tracking server is like a database that stores all our experiment data
# We tell MLflow where to find it (localhost:5000 means our local machine on port 5000)
# 
# Make sure the MLflow server is running in another terminal!
# Terminal command: mlflow server --host 127.0.0.1 --port 5000
#
# After running this, you can view the UI at: http://127.0.0.1:5000

mlflow.set_tracking_uri("http://127.0.0.1:5000")

print("✅ Connected to MLflow Tracking Server!")
print("📊 View the dashboard at: http://127.0.0.1:5000")
```

```python
# ============================================================================
# STEP 3: Create an Experiment and Log Your First Run
# ============================================================================
# An experiment is like a project folder that groups related training attempts
# Each training attempt is called a "run"
#
# mlflow.set_experiment(): Creates or selects an experiment
# mlflow.start_run(): Starts a new run (one training attempt)
# mlflow.log_metric(): Records a number (like accuracy, loss, etc.)
#
# The "with" statement automatically closes the run when done (best practice!)

# Create a new experiment called "check localhost connection"
mlflow.set_experiment("check localhost connection")

print("ℹ️  Experiment created: 'check localhost connection'")
print("📝 Starting first run...")

# Start a new run and log some test metrics
with mlflow.start_run():
    # Log two dummy metrics to test the connection
    # Metric names are user-defined (you can call them whatever makes sense)
    mlflow.log_metric("test", 1)           # Log a metric named "test" with value 1
    mlflow.log_metric("naveen", 2)         # Log a metric named "naveen" with value 2
    
print("✅ First run completed!")
```

```python
# ============================================================================
# Run 2: Log Another Run with Different Metrics
# ============================================================================
# We're still in the same experiment, but now starting a different run
# MLflow automatically creates a new run ID, so they're tracked separately
# This is how you compare multiple training attempts!

print("📝 Starting second run...")

with mlflow.start_run():
    # These metrics have slightly different names (added "-1" suffix)
    # This helps distinguish them from the first run
    mlflow.log_metric("test-1", 1)
    mlflow.log_metric("naveen-1", 2)
    
print("✅ Second run completed!")
```

```python
# ============================================================================
# Run 3: A Third Run to Demonstrate Multiple Comparisons
# ============================================================================
# Now we have 3 runs in the same experiment
# You can compare all 3 in the MLflow UI dashboard to find patterns!

print("📝 Starting third run...")

with mlflow.start_run():
    # More metrics with "-2" suffix
    mlflow.log_metric("test-2", 1)
    mlflow.log_metric("naveen-2", 2)
    
print("✅ Third run completed!")
print("\n🎉 SUCCESS! You've created your first MLflow experiment with 3 runs!")
print("📊 Go to http://127.0.0.1:5000 to see all your logged runs")
```
